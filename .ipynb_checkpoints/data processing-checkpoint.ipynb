```python
#key = git url (https://github.com/tensorflow/tensorflow)
#value = [embedding, primary topic, sub topic]
embeddings_dict = np.load("../all_repos_to_embedding/embeddings/batch1.npy",allow_pickle=True).item()
```

```python
#[open alex id, git url, primary topic, subtopic]
batch = np.load("data/batch1.npy")
batch[0]
```

```python
citations = pd.read_csv("data/has_each_company.csv")
citations.set_index('id', inplace=True)
```

```python
citations.head()
```

```python
#new_dict
#key = "open alex id", value = [embedding, citation relationship, primary topic, subtopic]
'''
batch -> get "open alex id", "primary topic", "subtopic"
embeddings_dict -> get "embedding"
citations -> get "citation relationship"

first combine batch + embedding_dict
'''
new_dict = {}
for key in embeddings_dict.keys():
    for i in batch:
        if(key==i[1]):
            id_val = i[0]
            values = embeddings_dict[key]
            if id_val in citations.index:
                related_citations = citations.loc[id_val]
                new_dict[id_val] = [values[0], related_citations, values[1], values[2]]
            # If id_val is not found, skip this iteration
            else:
                print(key)
                continue
```

```python
def parse_citation_string(citation_str):
    # Remove brackets, quotes, and split by commas
    citation_str = citation_str.strip()  # Remove any leading/trailing whitespace
    citation_str = citation_str.replace('"', '').replace('[', '').replace(']', '')  # Remove quotes and brackets
    
    # Split by commas and return a list of citation IDs, ignoring any empty results
    citation_list = [citation_id.strip() for citation_id in citation_str.split(',') if citation_id.strip()]
    return citation_list
```

```python
new_dict = {}
for batch_num in range(1, 7):
    # Load the embeddings and batch files dynamically
    embeddings_dict = np.load(f"../all_repos_to_embedding/embeddings/batch{batch_num}.npy", allow_pickle=True).item()
    batch = np.load(f"data/batch{batch_num}.npy")

    # Process each key in the embeddings dictionary
    for key in embeddings_dict.keys():
        for i in batch:
            if key == i[1]:
                id_val = i[0]
                values = embeddings_dict[key]

                # Check if id_val exists in the citations DataFrame
                if id_val in citations.index:
                    related_citations = parse_citation_string(citations.loc[id_val]["cited_by"])
                    institution = citations.loc[id_val]["institutions"]
                    year = citations.loc[id_val]["year"]
                    new_dict[id_val] = [values[0], related_citations, values[1], values[2], key, institution, year]
                else:
                    #print(f"ID {id_val} for key {key} not found in citations")
                    continue
```
